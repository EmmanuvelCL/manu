```python
#find the unsold or nonparctipent players in 2021 auction
ipl[ipl['2021 Squad'].isnull()]
```

```python
ipl['2021 Squad']=ipl['2021 Squad'].fillna('not paticipated')
ipl['2021 Squad']
```

```python
ipl.isnull().sum()
```

```python
teams=ipl[ipl['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
#add a new column status
ipl['status']=ipl['Team'].replace(teams,'sold')
ipl
```

```python
#how many players parcticipate in 2022 ipl auction
ipl.shape[0]
```

```python
#how many type of players have participated
types=ipl['TYPE'].value_counts()
types.reset_index()
```

```python
plt.pie(types.values,labels=types.index,labeldistance=1.3,autopct='1%2f%%',shadow=True,startangle=60)
plt.title('role of players participated',fontsize=15)
plt.plot
```

```python
#players sold and unslod in bar graph
plt.figure(figsize=(10,5))
fig=sns.countplot(ipl['status'],palette=['red','blue'])
plt.xlabel('sold or unsold')
plt.title('number of players')
plt.title('sold or unsold players',fontsize=15)
plt.plot()
```

```python
ipl.groupby('status')['Player'].count()
```

```python
#total numbers of players bought by each team
plt.figure(figsize=(20,10))
fig=sns.countplot(ipl[ipl['Team']!='unsold']['Team'])
plt.xlabel('team names')
plt.ylabel('number of players')
plt.title('players bought by each teams',fontsize=12)
plt.xticks(rotation=70)
plt.plot()
```

```python
ipl.groupby('Team')['Player'].count()
```

```python
#dislpay palyer on each team based on type of players
plt.figure(figsize=(20,10))
fig=sns.countplot(ipl[ipl['Team']!='unsold']['Team'],hue=ipl['TYPE'])
plt.title('players in each teams')
plt.xlabel('team names')
plt.ylabel('number of players')
plt.xticks(rotation=50)
```
